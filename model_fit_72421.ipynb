```python
import pandas as pd
import numpy as np
import os
import functions as f
import functions1 as f1
from scipy.special import kv as K_v  # Bessel function K_v
from statsmodels.stats.stattools import jarque_bera
from statsmodels.stats.diagnostic import acorr_ljungbox, het_arch
from scipy.stats import anderson_ksamp, PermutationMethod


# ==== 1. Đọc danh sách Station và tạo mapping ====
station_list_path = r'Station List Meteostat.xlsx'
data_dir = r'Weather Database Meteostat 2'

station_df = pd.read_excel(station_list_path, dtype={'Station ID': str})
station_map = dict(zip(station_df['Station ID'], station_df['City']))

# ==== 2. Chọn Station và danh sách khung thời gian ====
station_id = "72421"  # Ví dụ 1 Station
time_windows = [
    ("2020-07-01", "2024-06-30"),
    ("2020-08-01", "2024-07-31"),
    ("2020-09-01", "2024-08-31"),
    ("2020-10-01", "2024-09-30"),
    ("2020-11-01", "2024-10-31"),
    ("2020-12-01", "2024-11-30"),
]

city_name = station_map.get(station_id, "UnknownCity")

# ==== 3. Hàm thu thập tham số (giữ nguyên phần GH) ====
def collect_model_params(station_id, city_name, lambda_params, ar_params, sig_params, gh, start_date, end_date):
    a0 = a1 = a3 = a4 = np.nan
    omega_l = np.nan
    if lambda_params:
        coefs = lambda_params.get('coefs')
        if coefs is not None and len(coefs) >= 4:
            a0, a1, a3, a4 = coefs[:4]
        omega_l = lambda_params.get('omega', np.nan)
    a2 = np.sqrt(a3**2 + a4**2) if np.isfinite(a3) and np.isfinite(a4) else np.nan
    phi = np.arctan2(a3, a4) if np.isfinite(a3) and np.isfinite(a4) else np.nan

    beta1 = beta2 = beta3 = np.nan
    if ar_params:
        ar_coefs = ar_params.get('ar_coefs')
        if ar_coefs is not None and len(ar_coefs) >= 3:
            beta1, beta2, beta3 = ar_coefs[:3]

    if all(np.isfinite([beta1, beta2, beta3])):
        b1, b2, b3 = beta1, beta2, beta3
        alpha1 = 3 - b1
        alpha2 = 2 * alpha1 - b2 - 3
        alpha3 = alpha2 + 1 - alpha1 - b3
    else:
        alpha1 = alpha2 = alpha3 = np.nan

    # ==== Kiểm tra eigenvalue companion matrix ====
    eigvals = [np.nan, np.nan, np.nan]
    eig_check = "No"

    if all(np.isfinite([alpha1, alpha2, alpha3])):
        A = np.array([
            [0, 1, 0],
            [0, 0, 1],
            [-alpha3, -alpha2, -alpha1]
        ], dtype=float)
        try:
            eigvals_complex = np.linalg.eigvals(A)
            eigvals = eigvals_complex
            if np.all(np.real(eigvals_complex) < 0):
                eig_check = "Yes"
            print(f"Eigenvalues for window {start_date} - {end_date}: {eigvals_complex}")
        except Exception as e:
            print(f"[!] Lỗi tính eigenvalue: {e}")

    c = [np.nan] * 9
    if sig_params:
        sig_res = sig_params.get('sig_res')
        if sig_res is not None:
            params = np.asarray(sig_res.params)
            for i in range(min(len(params), 9)):
                c[i] = params[i]

    # ==== GH (giữ nguyên) ====
    gh_lambda = gh_alpha_bar = gh_mu = gh_sigma = gh_gamma = np.nan
    if gh and isinstance(gh, dict):
        ghp = gh.get('params', {}) or {}
        gh_lambda = ghp.get('lambda', np.nan)
        gh_alpha_bar = ghp.get('alpha.bar', np.nan)
        gh_mu = ghp.get('mu', np.nan)
        gh_sigma = ghp.get('sigma', np.nan)
        gh_gamma = ghp.get('gamma', np.nan)

    gh_chi = gh_psi = gh_alpha = gh_beta = gh_delta = np.nan
    if all(np.isfinite([gh_lambda, gh_alpha_bar])):
        try:
            K_lambda = K_v(gh_lambda, gh_alpha_bar)
            K_lambda_p1 = K_v(gh_lambda + 1, gh_alpha_bar)
            gh_psi = gh_alpha_bar * K_lambda_p1 / K_lambda
            gh_chi = gh_alpha_bar**2 / gh_psi
            gh_beta = gh_gamma / gh_sigma
            gh_delta = np.sqrt(gh_chi * gh_sigma)
            gh_alpha = np.sqrt(gh_sigma**-1 * (gh_psi + gh_gamma**2 / gh_sigma**2))
        except Exception:
            pass

    row = {
        'Station ID': station_id,
        'City': city_name,
        'Start': start_date,
        'End': end_date,
        'a0': a0, 'a1': a1, 'a2': a2, 'a3': a3, 'a4': a4, 'phi': phi,
        'beta1': beta1, 'beta2': beta2, 'beta3': beta3,
        'alpha1': alpha1, 'alpha2': alpha2, 'alpha3': alpha3,
        **{f'c{i+1}': c[i] for i in range(9)},
        'gh_lambda': gh_lambda, 'gh_alpha_bar': gh_alpha_bar,
        'gh_mu': gh_mu, 'gh_sigma': gh_sigma, 'gh_gamma': gh_gamma,
        'gh_chi': gh_chi, 'gh_psi': gh_psi,
        'gh_alpha': gh_alpha, 'gh_beta': gh_beta, 'gh_delta': gh_delta,

        # ==== Các cột eigenvalue ====
        'eig1': eigvals[0],
        'eig2': eigvals[1],
        'eig3': eigvals[2],
        'eig_check': eig_check,
    }

    return pd.DataFrame([row])


# ==== 4. Vòng lặp các khung thời gian (cập nhật) ====
all_params = []
eps_list = []  # Thu thập eps của tất cả window
file_path = os.path.join(data_dir, f"{station_id}.csv")

if not os.path.exists(file_path):
    print(f"[!] Không tìm thấy file cho station {station_id} ({city_name}).")
else:
    for start_date, end_date in time_windows:
        try:
            print(f"===> Xử lý {station_id} ({city_name}) từ {start_date} đến {end_date}...")
            df = f.preprocess(file_path, start=start_date, end=end_date)
            df2, lambda_params = f1.detrend_deseasonalize(df, T_col='T', title_city=city_name)
            
            # --- ADF test ---
            adf_result = f.adf_test_full(df2, title_city=city_name, show_output=True)
            
            # --- Fit AR & seasonal variance ---
            df3, ar_params = f1.fitCAR(df2, X_col='X', p=3, title_city=city_name)
            df4, sig_params = f1.seas_vol(df3, K=4, title_city=city_name)
            f.check_residuals(df4, title_city=city_name)
            
            # --- Thu thập eps để sau này chạy AD k-sample ---
            eps_clean = df4['eps'].dropna().values
            eps_list.append(eps_clean)
            
            # --- Residual checks ---
            jb_stat, jb_p, skew, kurt = jarque_bera(eps_clean)
            lb_eps = acorr_ljungbox(eps_clean, lags=[10], return_df=True)
            lb_eps2 = acorr_ljungbox(eps_clean**2, lags=[10], return_df=True)
            arch_stat, arch_p, f_stat, f_p = het_arch(eps_clean, nlags=10)

            resid_pvals = {
                'jb_p': jb_p,
                'lb_eps_p': lb_eps['lb_pvalue'].iloc[0],
                'lb_eps2_p': lb_eps2['lb_pvalue'].iloc[0],
                'arch_p': arch_p
            }

            # --- GH fit ---
            gh = f1.fit_gh(df4, col='eps', title_city=city_name)

            # --- Thu thập tất cả tham số + ADF + residual p-values ---
            df_row = collect_model_params(station_id, city_name, lambda_params, ar_params, sig_params, gh, start_date, end_date)
            df_row['adf_result'] = adf_result
            for k, v in resid_pvals.items():
                df_row[k] = v

            all_params.append(df_row)

        except Exception as e:
            print(f"[X] Lỗi trong khung thời gian {start_date}-{end_date}: {e}")


# ==== 5. Kiểm định AD k-sample giữa các window ====
ad_pvalue = np.nan
if len(eps_list) >= 2:
    # Chạy AD k-sample với permutation để p-value chính xác
    try:
        ad_res = anderson_ksamp(eps_list, method=PermutationMethod())
        ad_pvalue = ad_res.pvalue  # trả về giá trị đầy đủ, không bị cap
        print(f"AD k-sample p-value (permutation): {ad_pvalue:.4f}")
    except Exception as e:
        print(f"[X] Lỗi khi chạy AD k-sample với permutation: {e}")
else:
    print("[!] Không đủ window để chạy AD k-sample")


# ==== 6. Lưu kết quả ====
if all_params:
    param_df = pd.concat(all_params, ignore_index=True)
    param_df['ad_pvalue'] = ad_pvalue  # thêm cột AD p-value cho tất cả row
    out_path = fr'Weather_Model_Params_{city_name}_{time_windows[0][0]}_{time_windows[-1][1]}.xlsx'
    param_df.to_excel(out_path, index=False)
    print(f"\n✅ Đã lưu tham số cho {station_id} vào '{out_path}'")
else:
    print("⚠️ Không có dữ liệu nào được xử lý thành công.")
```
